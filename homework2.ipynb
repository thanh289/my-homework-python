```python
# Bài 2:Viết một hàm Python để tính toán giai thừa của một số (một số nguyên không âm).
#Gợi ý: Sử dụng hàm đệ quy n! = (n-1)!*n.
#Đệ quy là cách lập trình hoặc code một vấn đề, trong đó hàm tự gọi lại chính nó một hoặc nhiều lần trong khối code. Thông thường, nó trả giá trị trả về của lần gọi hàm. Nếu định nghĩa hàm thỏa mãn điều kiện đệ quy thì hàm được gọi là hàm đệ quy.
def tinh_giai_thua(n):
    if n==0 or n==1:
        return 1
    else:
        return n*(tinh_giai_thua(n-1))
n=int(input('Nhập số nguyên không âm bất kì:'))
print('Giai thừa của',n,'là',tinh_giai_thua(n))

```

```python
#Bài 4: Giả sử muốn tính tổng các số trong một chuỗi bất kì:
n=int(input("nhập số phần tử của chuỗi:"))  
list1=[]
for i in range(n):
    list1.append(int(input('các phần tử là')))
print(list1)
    
answer=0
for j in list1:
    answer+=j
print('tổng các phần tử là:',answer)
```

```python
# Bài 5:Nhập vào từ bàn phím số n và list cho trước, tìm các từ có độ dài lớn hơn n từ list đó.
n=int(input('Nhập n='))
my_list=list(input('Nhập list:').split())
print(my_list)

y=len(my_list)
for i in range(y):
    z=len(my_list[i])
    if z>n: 
        print ('Từ có độ dài lớn hơn',n,'là',my_list[i])
```

```python
#Bài 3: Viết một hàm Python chấp nhận một chuỗi và tính toán số chữ hoa và chữ thường.
a=input('Nhập chuỗi:')
x=0
y=0
for i in a:
    if i.islower():
        x+=1
    if i.isupper():
        y+=1 
print('Số kí tự in hoa là:', y)
print('Số kí tự chữ thường là:',x)
```

```python
#Bài 3: có sử dụng function:
def show(a):
    x = 0
    y = 0
    for i in a:
        if i.isupper():
            x += 1
        if i.islower():
            y += 1

    print("Cho chuỗi:", a)
    print("Số kí tự in hoa là:", x)
    print("Số kí tự chứ thường là:", y)


s = str(input())
show(s)
```

```python
# Bài 1. Viết một hàm số python có tham số đầu vào là 1 danh sách và các giá trị muốn xóa khỏi danh sách đó. Kết quả đầu ra là 1 danh sách đã được xóa đi các giá trị nhập vào. 
n=int(input("Nhập số phần tử của chuỗi:"))  
list1=[]
for i in range(n):
    list1.append(int(input('Các phần tử là')))
print(list1)

b=int(input('Nhập giá trị muốn xóa: '))
for i in list1:
    if b==i and list1.count(b)>1:
        list1.remove(b)
        print(list1)
    if b==i and list1.count(b)==1:
        list1.remove(b)
        print(list1)
```
