```python
# bài 3:Nhập vào số nguyên dương n từ bàn phím. Kiểm tra xem số n có phải là số nguyên tốt hay không. Nếu là số nguyên tố thì in ra dòng text : “Đây là số nguyên tố”. Nếu không thì in ra: “Không phải là số nguyên tố”. 

#CÁCH 1:
n=int(input())
đếmước=0
for i in range(1,n+1):
    if n%i==0:
        đếmước = đếmước +1
if đếmước ==2:
    print("Đây là số nguyên tố")
else:
    print("Không phải là số nguyên tố")
    
    
```

```python
#Bài 4: Nhập vào từ bàn phím số nguyên n. In ra tổng của các số thỏa mã hai điều kiện: nhỏ hơn n và là số chẵn.

n = int(input("Nhập vào một số nguyên bất kì:"))
s = 0
for i in range(2, n, 2):
    s += i
print('Tổng các số chẵn nhỏ hơn', n ,'là:', s)
```

```python
#Bài 8:Viết một chương trình chấp nhận đầu vào là một chuỗi các từ tách biệt bởi khoảng trắng, loại bỏ các từ trùng lặp rồi in chúng.

a=input("Nhập một chuỗi:")
b= a.split()
def cau_moi(b):
    chuỗi=''
    for từ in b:    
        if từ not in chuỗi:   
           chuỗi = chuỗi + " " + từ
    return chuỗi
print(cau_moi(b))
```

```python
#Bài 9: Yêu cầu người dùng cung cấp một chuỗi và cho biết đó có phải một palindrome không
def answer(s):
    if len(s)%2==1:
        if s[:len(s)//2] == s[:len(s)//2:-1]:
            print('Yes')
        else:
            print('No')
    if len(s)%2==0:
        if s[:len(s)//2] == s[:len(s)//2-1:-1]:
            print('Yes')
        else:
            print('No')
    
    
a = str(input('Nhập chuỗi:'))
answer(a)
```

```python
#Bài 10:Viết một chương trình chấp nhận chuỗi từ được phân tách bằng khoảng trống và in các từ chỉ gồm chữ số.
import re

string = input("nhập câu lệnh vào: ")
pattern = '\d+'

# Phương thức re.findall() trả về một danh sách các chuỗi chứa tất cả các kết quả khớp với pattern đưa ra.
result = re.findall(pattern, string) 
print(result)

```

```python
#Bài 11:
def Roman_to_Int(s):
    answer = 0
    value = {"I":1, "V":5, "X":10, "L":50, "C":100, "D":500, "M":1000}
    
    for i in range(len(s)):
        if i==0:
            answer = answer +value[s[i]]
        else:
            if value[s[i]] > value[s[i-1]]:
                answer = answer + value[s[i]] - 2*value[s[i-1]]
            else:
                answer = answer + value[s[i]]
    return answer
                
                
a=input('Nhập số la mã:')
print("Giá trị số nguyên tương ứng:", Roman_to_Int(a))
```
