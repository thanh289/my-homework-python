```python
# Bài 2:Viết một hàm Python để tính toán giai thừa của một số (một số nguyên không âm).
#Gợi ý: Sử dụng hàm đệ quy n! = (n-1)!*n.
#Đệ quy là cách lập trình hoặc code một vấn đề, trong đó hàm tự gọi lại chính nó một hoặc nhiều lần trong khối code. Thông thường, nó trả giá trị trả về của lần gọi hàm. Nếu định nghĩa hàm thỏa mãn điều kiện đệ quy thì hàm được gọi là hàm đệ quy.
def tinh_giai_thua(n):
    if n==0 or n==1:
        return 1
    else:
        return n*(tinh_giai_thua(n-1))
n=int(input('Nhập số nguyên không âm bất kì:'))
print('Giai thừa của',n,'là',tinh_giai_thua(n))

```

```python
#Bài 3: Viết một hàm Python chấp nhận một chuỗi và tính toán số chữ hoa và chữ thường.
#Cách 1:
def show(a):
    x = 0
    y = 0
    for i in a:
        if i.isupper():
            x += 1
        if i.islower():
            y += 1
    print("Số kí tự in hoa là:", x)
    print("Số kí tự chứ thường là:", y)
    
s = str(input("Nhập chuỗi: "))
show(s)
```

```python
#Cách 2:
def count_digit(s = "The Brow Fox"): 
    cnt_upper = 0
    cnt_lower = 0
    for i in s: 
        if i >'A' and i <'Z': 
            cnt_upper += 1 
        elif i>'a' and i < 'z':
            cnt_lower += 1 
    return cnt_upper, cnt_lower 

s=str(input('Nhập chuỗi:'))
print('Số kí tự in hoa và số kí tự chữ thường lần lượt là: ',count_digit(s))
```

```python
# PROBLEM 2: THREE INDICES

#hàm này tìm 3 số i, j , k tm đề bài
def find(lst=[]):
    n=len(lst)
    for i in range (0,n-2):
        for j in range (i+1,n-1):
            for k in range (j+1,n):
                if lst[i] < lst[j] and lst[j] > lst[k]:
                    return True, i, j, k
    return False , 0, 0, 0 #như kiểu để lấp đầy chỗ trống, cân bằng với các giá trị của return phía trên
    
    

    
if __name__ == "__main__":
    test_case=int(input("Nhập số lượng test case: "))
    while test_case != 0:
        n=int(input('số số nguyên trong dãy: '))
        tmp=list(input('hoán vị của dãy:').split())
        
        valid,i,j,k= find(tmp)
        if valid == True:
            print('YES')
            print(i+1,j+1,k+1)
            
        else:
            print('NO')
        test_case -= 1
```

```python
#hàm tạo cho 3 chuỗi có độ dài bằng nhau
def compare(a, b):
    h = max(len(a), len(b)) - min(len(a), len(b))
    if len(a)>len(b):
        return a, '0'*h + b
    else:
        return '0'*h + a, b
    
#hàm xóa đi số 0 đằng trước
def erase_0(wrong):
    str_=wrong[::-1]
    index = 0
    for i, j in enumerate(str_): 
        if j !='0':
            index = i 
            break 
    return str_[index:]

    

def sum_(a, b):
    ans = ''
    save = 0
    n=len(a)
    
    for i in range (0,n):
        value =( int(a[n-i-1]) +int(b[n-i-1]) + save ) %10
        ans += str(value)
        save=( int(a[len(a)-i-1]) +int(b[len(b)-i-1]) + save ) // 10
        
    if save != 0:
        ans += str(save)
        
    return erase_0(ans)
        
a='000000000000000099'
b='0001'
a,b=compare(a,b)
print(sum_(a,b))
    

```
