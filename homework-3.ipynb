```python
#BÀI 1: (MATH) Viết chương trình tính số tiền thực của một tài khoản ngân hàng dựa trên nhật ký giao dịch được nhập vào từ giao diện điều khiển.  Định dạng nhật ký được hiển thị như sau:
# D 100 ( D là tiền gửi). 
# W 200 ( W là tiền rút ra). 
amount=0
while True:
    giao_dich = input('Nhập giao dịch:')
    if not giao_dich:
        break
    m = giao_dich.split(' ')
    
    key= m[0]
    value= int(m[1])
    if key == 'D':
        amount += value
    elif key == 'W':
        amount -= value
    else:
        pass
print (amount)
        
```

```python
#BÀI 2: (MATH) DÃY FIBONACCI
def f(n):
    if n==0:
        return 0
    if n==1:
        return 1
    if n>1:
        return f(n-1) + f(n-2)
print(f(7))
```

```python
#BÀI 2: (MATH) 
#CÁCH 2:
def f(n):
    f0 = 0;
    f1 = 1;
    fn = 1;
 
    if (n == 0 or n == 1):
        return n;
    else:
        for i in range(2, n):
            f0 = f1;
            f1 = fn;
            fn = f0 + f1;
        return fn;
print(f(50))
```

```python
#BÀI 2: (DICTIONARY) Định nghĩa một hàm có thể in dictionary chứa key là các số từ 1 đến 10 và values là các giá trị bình phương của chúng.
def dic(n):
    d=dict()
    for i in range (1,n+1):
        d[i]=i**2
    return d
print(dic(10))
```

```python
#BÀI 1: (SET)Cô giáo Thảo thống kê chiều cao của các học sinh trong lớp 12A được kết quả như sau:
#161 182 161 154 176 170 167 171 170 174 150 142 148 165 170 178 156 145 149 163 162 159 165 165 170 180 155 159 155 153 152 162 180 168 169 168 167 170
#a. Hỏi lớp có bao nhiêu học sinh?
#b. Tính chiều cao trung bình của các học sinh trong lớp.
#c. Liệt kê các chiều cao khác nhau của học sinh trong lớp. Tính giá trị trung bình của chúng.

from math import *

lst_height=[161, 182, 161, 154, 176, 170, 167, 171, 170, 174, 150, 142, 148, 165, 170, 178, 156, 145, 149, 163,
            162, 159, 165, 165, 170, 180, 155, 159, 155, 153, 152, 162, 180, 168, 169, 168, 167, 170]
std = len(lst_height)
print('Lớp có',std,'học sinh')

tong1 = 0
for i in lst_height:
    tong1 += i
    
TB1 = tong1/std
print('Chiều cao trung bình các học sinh trong lớp là:',round(TB1,2))

s_height = set(lst_height)
print('Các chiều cao khác nhau của học sinh trong lớp:',s_height)

hei=len(s_height)
tong2 = 0
for i in s_height:
    tong2 += i
TB2 = tong2/hei
print('Giá trị trung bình của chúng là:',TB2)
```

```python
#BÀI 2: (TUPLE) Cho một dãy số nguyên từ 1 đến n. Sử dụng tuple hiển thị ra màn hình nửa dãy đầu tăng dần,
#nửa dãy sau giảm dần. 
lst=[]
n=int(input("Nhập n:"))
for i in range (1,n+1):
    lst.append(i)
tule=tuple(lst)
print('Ta có dãy:',tule)

a=tule[:n//2]
print('Nửa dãy đầu tăng dần:',a)

b=tule[n//2:n+1]
c=sorted(b,reverse=True)
print('Nửa dãy sau giảm dần:',tuple(c))
```
