```python
# BÀI 2: Cho một chuỗi string s chỉ bao gồm các ký tự:  '(', ')', '{', '}', '[' và ']', Xác định nếu chuỗi s là hợp lệ. 
# Một chuỗi được coi là hợp lệ nếu:
# 1.Dấu ngoặc mở phải được đi cùng với dấu ngoặc đóng cùng loại.
# 2.Dấu ngoặc phải được đóng theo đúng thứ tự.

#chuỗi chắc chắn sẽ sai nếu dấu đầu tiên là đóng
#vì bài này liên quan đến thứ tự, vậy nên sử dụng stack
#ý tưởng là tạo một list rỗng
#quét các phần tử có trong chuỗi mình nhập
#nếu nó là dấu mở thì append vào list rỗng
#còn nếu gặp dấu đóng thì xóa đi dấu mở ngoài cùng có trong list vì nó đã tạo thành một cặp
#(đương nhiên vì đang nghĩ đến stack, mà muốn xóa phần tử ta sẽ dùng pop())
#chuỗi sẽ đúng nếu list của bạn dc xóa sạch 


def isvalid(s: str) -> bool:
    #khởi tạo một list rỗng:
    stack=[]
    #tạo một dict chứa key là dấu đóng, value là dấu mở:
    dic={'}':'{', ')':'(' ,']':'['}
    
    for i in s:
        #nếu i là dấu mở:
        if i not in dic:
            stack.append(i)
        #còn nếu là dấu đóng:
        else:
            #nếu độ dài list >0 và phần tử ngoài cùng của dic giống với dấu đóng đang xét thì xóa đi
            if len(stack)>0 and stack[-1]==dic[i]:
                stack.pop()
            else:
                #nếu như đã xóa đi hết các ngoặc t/m (hoặc ngay từ đầu) trong chuỗi chỉ còn dấu đóng, lệnh này sẽ ngay được thực thi
                return False

isvalid('((()))')
            
            
```

```python
# BÀI 1: Cho một chuỗi string s chỉ bao gồm các toán tử và toán hạng. 
#Trong đó:   Toán tử chỉ bao gồm các phép: + - * /
#            Toán hạng là các số nguyên a ( 0<=0 <=999).
#Nhiệm vụ của bạn là tính giá trị của biểu thức đã cho. 


#HÀM XÁC ĐỊNH ĐỘ ƯU TIÊN CỦA DẤU
def priority(op): 
    if op == '+':
        return 0
    elif op == '-': 
        return 1
    else:
        return 2
    
    
    
#HÀM CỘNG TRỪ CƠ BẢN
def process_op(val, op): 
    r = val.pop()
    l = val.pop() 
    if op == '+':
        val.append(l+r)
    elif op == '-':
        val.append(l-r)
    elif op == '*': 
        val.append(l*r)
    elif op == '/': 
        val.append(l/r)
        
    return val 



#HÀM XÁC ĐỊNH TOÁN HẠNG
def isdigit (c): 
    valid = "0123456789"
    return valid.find(c)!= -1 


#HÀM TÍNH CHUẨN
def evaluate(s): 
    val = [] #list chứa toán hạng
    op  = [] #list chứa toán tử
    
    # BƯỚC 1
    for i in s: 
        #nếu là toán tử append vào val
        if isdigit(i) is True: 
            val.append(int(i))
        else: 
            #nếu dấu ngoài cùng đang có trong ngoặc độ ưu tiên lớn hơn dấu đang xét thì thực hiện phép tính luôn
            while len(op)!=0 and priority(op[-1]) >= priority(i): 
                val = process_op(val, op[-1]) 
                #thực hiện xong thì xóa dấu đó đi
                op.pop()
            #nếu hàm while k thực hiện thì append dấu đang xét vào luôn, còn nếu while đúng thì phải thực hiện tính xong thì mới append vào
            op.append(i) 
    # BƯỚC 2

    while len(op)!= 0: 
        #hiện tại các dấu đang sắp xếp có độ ưu tiên tăng dần
        val = process_op(val,op[-1]) 
        op.pop()

    return val[-1]

print(evaluate('1+2+3'))
```
